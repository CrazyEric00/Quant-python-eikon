```python
def cov(ri,rm):
    ans=0
    avg=0
    for i in range(len(ri)):
        avg+=np.float64(ri[i])
    avg/=10
    for i in range(len(ri)):
        ans+=((ri[i]-avg)*(rm[i]-_rm)/(10-1))
    return ans

def cov_r(ri,rj):
    ans=0
    avg1=0
    avg2=0
    for i in range(len(ri)):
        avg1+=np.float64(ri[i])
        avg2+=np.float64(rj[i])
    avg1/=10
    avg2/=10
    for i in range(len(ri)):
        ans+=(ri[i]-avg1)*(rj[i]-avg2)/(10-1)
    return ans
```
