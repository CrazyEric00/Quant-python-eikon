```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel('value.xls')
data
```

```python
mt=data.iloc[:4,:]
yt=data.iloc[4:11,:]
yt
```

```python
def name_to_digits(Name):
    ans=0
    for i in range(len(Name)):
        if Name[i]>='0' and Name[i]<='9':
            ans=ans*10+int(Name[i])
    return ans
```

```python
def cal_mt(Name,Yield,Month):
    price=100/((1+Yield*0.01/12*Month)**(Month/12))
    return price
```

```python
def cal_yt(Name,Coupon,Yield,Year):
    price=0
    for i in range(1,Year):
        price+=Coupon/(1+Yield*0.01)**i
    price+=(Coupon+100)/(1+Yield*0.01)**Year
    return price
```

```python
ans=[]
```

```python
for i in range(len(mt)):
    a=list(mt.iloc[i].values)
    Name=a[0]
    Yield=a[4]
    Yield=np.float64(Yield)
    Month=name_to_digits(Name)
    ans.append([Name,cal_mt(Name,Yield,Month)])
```

```python
for i in range(len(yt)):
    a=list(yt.iloc[i].values)
    Name=a[0]
    Coupon,_=a[1].split('/')
    Coupon=np.float64(Coupon)
    Yield=a[4]
    Yield=np.float64(Yield)
    Year=name_to_digits(Name)
    ans.append([Name,cal_yt(Name,Coupon,Yield,Year)])
```

```python
df=pd.DataFrame(ans,columns=['Name','The estimate'])
df
```

```python
df.to_excel('estimate.xls',index=False)
```
